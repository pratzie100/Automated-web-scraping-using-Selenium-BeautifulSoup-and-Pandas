```python
import pandas as pd
import re
columns = ['Name', 'Shoe URL', '4.5', '5', '15', '43', '12', '3', '41', '11', '14',
       '8', '1', '13.5', '10.5', '5.5', '13', '9', '1.5', '12.5', '19', '8.5',
       '4', '17', '16', '18', '17.5', '2.5', '6', '40', '42', '39', '15.5',
       '10', '6.5', '3.5', '16.5', '18.5', '14.5', '11.5', '9.5', '45', '7',
       '2', '7.5', '38', '44']



# Function to extract numeric part of shoe size
def extract_shoe_size(column_name):
    # Use regex to find numeric part of the shoe size
    match = re.search(r'\d+(\.\d+)?', column_name)
    if match:
        return float(match.group())
    else:
        return float('inf')  # Return infinity for non-numeric columns

# Sort columns based on shoe size
sorted_columns = sorted(columns[2:], key=extract_shoe_size)

# Rearrange columns in DataFrame
sorted_columns = ['Name', 'Shoe URL'] + sorted_columns  # Place 'Name' and 'Url' at the beginning
temp = temp[sorted_columns]
```
